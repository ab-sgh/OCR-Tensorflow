Importing relevant Libraries

```python
import tensorflow as tf 
from keras.models import Sequential
from keras.layers import Conv2D, Flatten, MaxPooling2D, Dense
import os
import cv2
import numpy as np
```

Preprocessing Image data

```python
import os
import cv2
import numpy as np

images = []
labels = []

path = 'data/training_data'

dir_list = os.listdir(path)
for i in dir_list:
    dir = os.path.join(path, i)
    file_list = os.listdir(dir)
    for j in file_list:
        files = os.path.join(dir, j)
        img = cv2.imread(files)
        
        # Resize the image to (64,64)
        img = cv2.resize(img, (64, 64))
        
        # Apply Gaussian Blur for noise reduction
        img = cv2.GaussianBlur(img, (5, 5), 0)
        
        # Normalize the pixel values
        img = np.array(img, dtype=np.float32)
        img = img / 255.0
        
        # Append the processed image and label
        images.append(img)
        labels.append(i)
```

```python
X = np.array(images)
y = np.array(labels)
```

Encode labels in y to numeric values

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.utils import shuffle
X_sh, y_sh = shuffle(X, y, random_state =42)
```

Convolution Layer

```python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(3,3), activation='relu', input_shape=(64,64,3)))
model.add(MaxPooling2D())
model.add(Conv2D(filters=32, kernel_size=(3,3),  activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(filters=64, kernel_size=(3,3),  activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
model.add(Flatten())
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=36, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics = ['accuracy'])
history = model.fit(X_sh, y_sh ,validation_split=0.2, batch_size=16, epochs=10)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
```

Pre-processing Testing Data

```python
test_images = []
test_labels = []
 
path = 'data/testing_data'
 
dir_list = os.listdir(path)
for i in dir_list:
  dir = os.path.join(path, i)
  file_list = os.listdir(dir)
  for j in file_list:
    files = os.path.join(dir, j)
    img = cv2.imread(files)
    img = cv2.resize(img, (64,64))
    img = np.array(img, dtype=np.float32)
    img = img/255
    test_images.append(img)
    test_labels.append(i)
```

```python
X_test = np.array(test_images)
y_test = np.array(test_labels)
```

```python
preds = model.predict(X_test)
predicted_labels = le.inverse_transform(np.argmax(preds, axis=1))
```

Test Accuracy

```python
y_test = le.fit_transform(y_test)
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
```
